```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('E:\\project csv files\\sample_superstore\\SampleSuperstorecleaned.csv')
```

```python
df.head(20)
```

```python
#Total sales and profit
totalSales=df['Sales'].sum()
totalProfit=df['Profit'].sum()
print(f'TotalSales :{totalSales}')
print(f'TotalProfit :{totalProfit}')
```

```python
#sales distribution Vs Profit Distribution
fig,ax=plt.subplots(1,2,figsize=(14,5))
ax[0].hist(df["Sales"].dropna(), bins=40)
ax[0].set_title("Sales Distribution")
ax[0].set_xlabel("Sales")
ax[0].set_ylabel("Frequency")
ax[1].hist(df['Profit'].dropna(),bins=40)
ax[1].axvline(0)
ax[1].set_title("Profit Distribution")
ax[1].set_xlabel("Profit")
ax[1].set_ylabel("Frequency")
```

```python
a = df.groupby('Category')
a.groups
```

```python
for category,group in df.groupby('Category'):
    plt.scatter(group['Sales'], group['Profit'], label=category)
plt.xlabel("Sales")
plt.ylabel("Profit")
plt.title("Sales vs Profit by Category")
plt.legend()
plt.show()    
```

```python
#Region - wise profit and sales
import matplotlib.pyplot as plt
Region_sales_profit=df.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
print(Region_sales_profit)
plt.figure(figsize=(10,6))
Region_sales_profit.plot(kind='bar', figsize=(10,6),color=['lightblue','lightgreen'])
plt.title ('SalesandProfit by Region')
plt.ylabel('Values')
plt.grid(axis='x')
plt.show()
```

```python
#category- wise profit and sales
import matplotlib.pyplot as plt
Categorywise_SalesProfit=df.groupby('Category')[['Sales','Profit']].sum()
print(Categorywise_SalesProfit)
plt.figure(figsize=(8,6))
Categorywise_SalesProfit.plot(kind='bar',figsize=(8,6))
plt.title('Sales vs Profit by Category')
plt.ylabel('Values')
plt.show()
```

```python
#sub-category- wise profit and sales
import matplotlib.pyplot as plt
Categorywise_SalesProfit=df.groupby('Sub-Category')[['Sales','Profit']].sum()
print(Categorywise_SalesProfit)
plt.figure(figsize=(8,6))
Categorywise_SalesProfit.plot(kind='barh',figsize=(8,6))
plt.title('CategoryWise profit VS sales')
plt.xlabel('Values')
plt.ylabel('Sub_Category')
plt.grid(axis='x')
plt.show()
```

```python
#Profit by Sub-Category
from matplotlib.pyplot import colormaps
subcategory_profit=df.groupby('Sub-Category')['Profit'].sum()
plt.figure(figsize=(8,6))
subcategory_profit.plot(kind='barh',figsize=(8,6))
plt.get_cmap('Pastel1')
plt.title('Profit by Sub-Category')
plt.xlabel('Profit')
plt.ylabel('Sub_category')
plt.grid('True')
plt.show()
```

```python
#Category wise profit
Profitbycategory=df.groupby('Category')['Profit'].sum()
plt.figure(figsize=(8,4))
Profitbycategory.plot(kind='bar',x='Category',y='Profit',figsize=(8,6),color=['tomato', 'gold', 'palegreen'])
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
```
